```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from tkinter import *
from tkinter import messagebox

dataset_url = r"C:\heart_ds\heart.csv"
dataset_df = pd.read_csv(dataset_url)
dataset_df.head()
```

```python
X = dataset_df.iloc[:, :-1].values
y = dataset_df.iloc[:, -1].values
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size = 0.2 , random_state = 0 )
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=2 , metric = 'minkowski', p = 2)
knn.fit(X_train, y_train)
```

```python
print(knn.predict(sc.transform([[53, 1, 0, 140, 203, 1, 0, 155, 1, 3.1, 0, 0, 3]])))
```

```python
y_pred = knn.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
#making confusion matrix
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
